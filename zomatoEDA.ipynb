```python
country_df = pd.read_excel(r"C:\Users\HP\Downloads\Country-Code.xlsx") 
country_df.head()
```

```python
#Merging the two datasets 
final_df = pd.merge(df,country_df,on='Country Code') 
final_df.head()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating_Rate'})
ratings
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x='Aggregate rating', y = 'Rating_Rate',data=ratings,hue='Rating color',palette=['White','Red','Orange','Yellow','Green','Green'])
```

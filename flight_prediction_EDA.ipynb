```python
#Importing the neccessary libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
#Importing the training dataset 
train_df = pd.read_excel(r"C:\Users\HP\Downloads\Data_Train.xlsx") 
```

```python
train_df.head()
```

```python
#importing the test data 
test_data = pd.read_excel(r"C:\Users\HP\Downloads\Test_set.xlsx") 
```

```python
test_data.head()
```

```python
#Now we can join the train and test data 
final_df = pd.concat([train_df,test_data])
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
#Checking how many null values are there in columns 
final_df.isnull().sum()
```

```python
#WE can apply the feature engineering on the 'Date_of_Journey' 
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0] 
```

```python
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1] 
```

```python
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2] 
```

```python
final_df.info()
```

```python
#Changing the datatypes of the 'Date','Month' and 'Year' 
final_df['Date'] = final_df['Date'].astype(int) 
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int) 
```

```python
final_df.info()
```

```python
#Now we can drop the 'Date_of_Journey' column 
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
#Now lets move to the 'Arrival_TIme' column 
final_df['Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['arrival_hour'] = final_df['Time'].str.split(':').str[0] 
final_df['arrival_hour']
```

```python
final_df['Arrival_minutes'] = final_df['Time'].str.split(':').str[1] 
final_df['Arrival_minutes']
```

```python
final_df.info()
```

```python
final_df.drop('Time',axis=1,inplace=True) 
```

```python
final_df['arrival_hour'] = final_df['arrival_hour'].astype(int) 
final_df['Arrival_minutes'] = final_df['Arrival_minutes'].astype(int)  
```

```python
final_df.info()
```

```python
#Now we can drop the Arrival_Time column 
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
#Mapping the 'Total_Stops' values to numerical values 
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':2}) 
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
#Dropping the Route column 
final_df.drop('Route',axis=1,inplace=True) 
```

```python
final_df.head()
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].str.split(':').str[0] 
final_df['Dep_Minute'] = final_df['Dep_Time'].str.split(':').str[1] 
```

```python
final_df.info()
```

```python
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int) 
final_df['Dep_Minute'] = final_df['Dep_Minute'].astype(int)
```

```python
final_df.head()
```

```python
#Droping the 'Dep_Time' column 
final_df.drop('Dep_Time',axis=1,inplace=True) 
```

```python
final_df.head(1)
```

```python
final_df['Duration_Hour'] = final_df['Duration'].str.split('h').str[0] 
final_df['Duration_Hour']
```

```python
final_df['Duration_minutes'] = final_df['Duration'].str.split('h').str[1] 
final_df['Duration_minutes'] = final_df['Duration_minutes'].str.replace('m','')
```

```python
final_df['Duration_minutes']
```

```python
final_df.head()
```

```python
#Dropping the 'Duration' column 
final_df.drop('Duration',axis=1,inplace=True) 
```

```python
final_df.head(1)
```

```python
#Now we can do the label encoding for the 'Airline','Source','Destination' and 'Additional_Info' 
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
```

```python
final_df['Airline']  = le.fit_transform(final_df['Airline']) 
```

```python
final_df['Source'] = le.fit_transform(final_df['Source']) 
final_df['Destination'] = le.fit_transform(final_df['Destination']) 
final_df['Additional_Info'] = le.fit_transform(final_df['Additional_Info'])
```
